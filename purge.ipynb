Remove non-wav data

```python
import os
PATH = "voice/dataset/32k/"
TARGET = "Ayaka/"
for file in os.listdir(f'{PATH}{TARGET}'):
    if file.split('.')[-1] != 'wav':
        os.remove(PATH+TARGET+file)
        print("removed: ", file)
```

Remove wav file

```python
for file in os.listdir(f'{PATH}{TARGET}'):
    if file.split('.')[-1] == 'wav':
        os.remove(PATH+TARGET+file)
        print("removed: ", file)
```
